```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
team_season = pd.read_csv('team_season.csv')
```

```python
team_season
```

```python
team_season.head()
```

```python
# 将比赛时间转换成赛季，当年6月之后的比赛为下一赛季
def convert_time_to_season(s):
    assert type(s) == str
    date = s.split('-')
    if int(date[1]) >= 7:
        return int(date[0]) + 1
    else:
        return int(date[0])
```

```python
team_season['赛季'] = team_season['时间'].map(convert_time_to_season)
team_season.head()
```

```python
# 统计每个赛季的得分
three_point_season = team_season.groupby(team_season['赛季']).mean()
three_point_season
```

```python
three_point_season[['三分出手', '三分命中']].plot(kind='bar')
plt.xlabel('Season')
```

```python
plt.show()
```

```python
# 抽取每个赛季的常规赛战绩十佳球队,查看其三分的场均出手数与场均命中数
three_point_attempt = []
three_point_hit = []

for year in range(1986, 2017):
    single_season = team_season[team_season['赛季'] == year]
    team_single_season_mean = single_season.groupby(single_season['球队']).mean()
    team_current_season_win = []
    for i in range(len(team_single_season_mean)):
        team = team_single_season_mean.iloc[i].name
        team_single_season_data = single_season[single_season['球队'] == team]
        team_single_season_win = team_single_season_data['结果'].value_counts().W
        team_current_season_win.append(team_single_season_win)
    team_single_season_mean['胜场数'] = team_current_season_win
    team_single_season_mean.sort_values(by='胜场数', ascending=False, inplace=True)
    team_single_season_mean = team_single_season_mean.head(10).mean()
    three_point_attempt.append(team_single_season_mean['三分出手'])
    three_point_hit.append(team_single_season_mean['三分命中'])

three_point_top10 = DataFrame({'三分出手':three_point_attempt, '三分命中':three_point_hit},index=np.arange(1986,2017))
three_point_top10
```

```python
three_point_top10.plot(kind='bar')
plt.xlabel('Seanson')
plt.show()
```

```python
# 抽取几个单赛季的联盟所有球队的数据，查看单赛季中三分命中率的变化趋势是否与该球队的胜率相符合 以2016为例子
team_season_16 = team_season[team_season['赛季'] == 2016]
team_season_16.head()
```

```python
team_season_16_data = team_season_16.groupby(team_season_16['球队']).sum()
team_season_16_data
```

```python
win_lose_16 = {}

for i in range(len(team_season_16)):
    match = team_season_16.iloc[i]
#     print(match)
    team = match['球队']
    outcome = match['结果']
    if team not in win_lose_16.keys():
        win_lose_16[team] = {}
    if outcome not in win_lose_16[team].keys():
        win_lose_16[team][outcome] = 1
    else:
        win_lose_16[team][outcome] += 1
        
win_lose_16 = DataFrame(win_lose_16).T
# print(win_lose_16)
win_lose_16['win_rate'] = win_lose_16['W'] / 82
win_lose_16.sort_values(by='win_rate', ascending=False)
```

```python
# 计算2016赛季所有球队三分命中率
three_point_to_win_16 = pd.merge(team_season_16_data[['三分出手', '三分命中']], win_lose_16, left_index=True, right_index=True)
```

```python
three_point_to_win_16
```

```python
three_point_to_win_16.rename(columns={'三分出手':'3PA', '三分命中':'3P'}, inplace=True)
three_point_to_win_16['3P%'] = three_point_to_win_16['3P'] / three_point_to_win_16['3PA']
three_point_to_win_16.sort_values(by='win_rate', inplace=True)
```

```python
# 绘图
plt.bar(np.arange(30), three_point_to_win_16['win_rate'], label='WinRate')
plt.bar(np.arange(30), -three_point_to_win_16['3P%'], label='3P%')
plt.xticks(np.arange(30), list(three_point_to_win_16.index), rotation=90)
plt.xlabel('Team')
plt.show()
```

```python
# 16赛季所有球队三分出手与三分命中情况
plt.bar(np.arange(30), three_point_to_win_16['3PA'])
plt.bar(np.arange(30), three_point_to_win_16['3P'])
plt.hlines(three_point_to_win_16['3PA'].mean(), 0, 30, linestyles='dashed')
plt.hlines(three_point_to_win_16['3P'].mean(), 0, 30, linestyles='dashed')
plt.xticks(np.arange(30), list(three_point_to_win_16.index), rotation=90)
plt.xlabel('Team')
plt.show()
```

```python
# 计算每个赛季的三分命中率排名与球队战绩排名之间这两个变量之间的相关系数，查看变化情况
coef = []

for year in range(1986, 2017):
    win_count = []
    win_index = []
#     每个赛季的数据
    single_season = team_season[team_season['赛季'] == year]
    team_sum = single_season.groupby(single_season['球队']).sum()
    team_sum['3P_Rate'] = team_sum['三分命中'] / team_sum['三分出手']
    team_sum.sort_values(by='3P_Rate', ascending=False, inplace=True)
#     单个赛季每个球队的数据
    for team in range(len(team_sum)):
        team_name = team_sum.iloc[team].name
        current_team_data = single_season[single_season['球队'] == team_name]
        win_count.append(current_team_data['结果'].value_counts().W)
    win_count.sort()
    win_count.reverse()
    print(win_count)
#     找出每支球队按胜场排列的排名
    for team in range(len(team_sum)):
        team_name = team_sum.iloc[team].name
        current_team_data = single_season[single_season['球队'] == team_name]
        current_win = current_team_data['结果'].value_counts().W 
        win_index.append(win_count.index(current_win))
    
#     print(win_index)
    team_sum['Win_Rank'] = win_index
    team_sum['3P_Rate_Rank'] = np.arange(len(team_sum))
    coef.append(team_sum['Win_Rank'].corr(team_sum['3P_Rate_Rank']))
coef
```

```python
plt.plot(coef)
plt.xticks(np.arange(31), np.arange(1986, 2017), rotation=90)
plt.xlabel('Year')
plt.ylabel('coef')
plt.title('coef between Win Rate and 3P%')
plt.show()
```
