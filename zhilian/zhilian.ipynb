```python
results = []
for i in range(0, 90, 90):
    global results
    url = 'https://fe-api.zhaopin.com/c/i/sou?start=%d&pageSize=90&cityId=538&salary=0,0&workExperience=-1&education=-1 \
    &companyType=-1&employmentType=-1&jobWelfareTag=-1&kw=python&kt=3' % (i)
    req = requests.get(url)
    content = req.json()
    results.append(content.get('data').get('results'))
    print(results)

        
```

```python
zhilian = []
for result in results:
    for one in result:
        data = []
        try:
            name = one.get('jobName')
            company = one.get('company').get('name')
            salary = one.get('salary')
            eduLevel = one.get('eduLevel').get('name')
            workingExp = one.get('workingExp').get('name')
            emplType = one.get('emplType')
            detail_url = one.get('positionURL')  
            data = [name, company, salary, eduLevel, workingExp, emplType, detail_url]
#             req = requests.get(detail_url, headers=headers)
#             if(req.status_code) == 200:
#                 print(req.cookies)
#             for cookie in req.cookies:
#                 print(cookie)
#             time.sleep(3)
        except Exception as e:
            print(e)
            continue
#         print(req.text)
#         soup = bs(req.text, 'lxml')
#         descrip = soup.find(class_='describtion__detail-content')
#         detail = descrip.find('p')
#         print(detail)
        zhilian.append(data)
    print(zhilian)
        
        
```

```python
with open('zhilian.csv', 'a+', newline='', encoding='utf-8-sig') as f:
    csvfile = csv.writer(f)
    for info in zhilian:
        csvfile.writerow(info)
```
